```python
#data analysis completed
#now the preprocessing work
Adj_close_price = google_data[['Close']]
max(Adj_close_price.values),min(Adj_close_price.values)
```

```python
#for getting data in 0 to 1 range.
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_data = scaler.fit_transform(Adj_close_price)
scaled_data
```

```python
len(scaled_data)
```

```python
x_data = []
y_data = []

for i in range(100, len(scaled_data)):
    x_data.append(scaled_data[i-100:i])
    y_data.append(scaled_data[i])


import numpy as np
x_data, y_data = np.array(x_data), np.array(y_data)

    
```

```python
x_data[0],y_data[0]
```

```python
int(len(x_data)*0.7)
```

```python
4908-100-int(len(x_data)*0.7)
```

```python
splitting_len = int(len(x_data)*0.8)
x_train = x_data[:splitting_len]
y_train = y_data[:splitting_len]  # ✅ Correct assignment

x_test = x_data[splitting_len:]
y_test = y_data[splitting_len:]
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
from keras.models import Sequential
from keras.layers import Dense, LSTM
```

```python
model = Sequential()
model.add(LSTM(128, return_sequences = True, input_shape=(x_train.shape[1],1)))
model.add(LSTM(64,return_sequences=False))
model.add(Dense(25))
model.add(Dense(1))
```

```python
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
model.fit(x_train, y_train, batch_size = 1, epochs = 2)
```

```python
model.summary()
```

```python
predictions = model.predict(x_test)
```

```python
inv_predictions = scaler.inverse_transform(predictions.reshape(-1, 1))
inv_predictions
```

```python
inv_y_test = scaler.inverse_transform(y_test.reshape(-1, 1))
inv_y_test
```

```python
rmse = np.sqrt(np.mean((inv_predictions.flatten() - inv_y_test.flatten())**2))
rmse
```

```python
ploting_data = pd.DataFrame(
    {
        'original_test_data': inv_y_test.reshape(-1),
        'predictions': inv_predictions.reshape(-1)
    },
        index = google_data.index[splitting_len+100:]
)
ploting_data.head()
```

```python
plot_graph((15,6), ploting_data, 'test data')
```

```python
plot_graph((15,6), pd.concat([Adj_close_price[:splitting_len+100],ploting_data], axis=0), 'whole data')
```
